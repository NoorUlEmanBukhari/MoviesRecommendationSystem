```python
import pandas as pd
import numpy as np
import ast
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from surprise import SVD, Dataset, Reader
from surprise.model_selection import train_test_split
import ast
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
# Step 2: Merge datasets
movies = movies.merge(credits, on='title')
```

```python
movies.head()
```

```python
# Step 3: Preprocess and Extract Important Features
def convert(obj):  
    """Convert genres and keywords into list of names."""
    L = []
    for i in ast.literal_eval(obj): #Transform JSON-like data (e.g., genres or keywords) into a list of names.
        L.append(i['name'])
    return L

movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def get_cast(obj):
    """Extract top 3 cast members."""
    L = []
    counter = 0
    for i in ast.literal_eval(obj): #Use ast.literal_eval(obj) to safely convert the JSON-like string into a Python list of dictionaries.
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L

movies['cast'] = movies['cast'].apply(get_cast)
```

```python
def get_director(obj):
    """Extract director from crew."""
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director': #Loop through the list and check the job field for "Director.movies are known for their unique storytelling and dialogue."
            L.append(i['name'])
            break
    return L

movies['crew'] = movies['crew'].apply(get_director)
```

```python
# Step 4: Clean and Combine Tags
movies['overview'] = movies['overview'].apply(lambda x: x.split() if isinstance(x, str) else []) #It splits the movie description (overview) into a list of words, so each word in the description becomes a separate item in a list.
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])

movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
new_df = movies[['movie_id', 'title', 'tags']]
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x)) #This converts the list into a single string by joining the elements of the list with a space " " between them.
```

```python
# Step 5: Content-Based Filtering
cv = CountVectorizer(max_features=5000, stop_words='english') #converts a list of words into a matrix of word counts.
vectors = cv.fit_transform(new_df['tags']).toarray()
similarity = cosine_similarity(vectors)
```

```python
# Function to recommend movies based on content similarity
def content_recommend(movie):
    if movie in new_df['title'].values:
        movie_index = new_df[new_df['title'] == movie].index[0]
        distances = similarity[movie_index]
        movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
        return [(new_df.iloc[i[0]].title, new_df.iloc[i[0]].movie_id) for i in movie_list]
    return []                                                                                # if a movie exists, calculates similarities with other movies, and returns the top 5 most similar movies. If the movie doesn't exist, it returns an empty list.
```

```python
# Step 6: Collaborative Filtering using Surprise Library
ratings = pd.read_csv('ratings.csv')  # The code loads ratings data, splits it, and trains a model (SVD) to predict movie ratings based on user behavior.

reader = Reader(rating_scale=(0.5, 5))
data = Dataset.load_from_df(ratings[['userId', 'movieId', 'rating']], reader)
trainset, testset = train_test_split(data, test_size=0.2)

algo = SVD()
algo.fit(trainset)
```

```python
# Function for collaborative filtering recommendation
def collaborative_recommend(user_id, n_recommendations=5):
    user_movies = ratings[ratings['userId'] == user_id]['movieId'].tolist()  #The function recommends 5 movies the user hasn't seen yet, based on predicted ratings.
    all_movies = movies['movie_id'].tolist()                                 
    unseen_movies = [m for m in all_movies if m not in user_movies]          

    predictions = [(movie, algo.predict(user_id, movie).est) for movie in unseen_movies]
    predictions = sorted(predictions, key=lambda x: x[1], reverse=True)[:n_recommendations]
    return [(new_df[new_df['movie_id'] == p[0]]['title'].values[0], p[1]) for p in predictions]
```

```python
# Step 7: Add New Movie if Not Found in Dataset
def add_new_movie(movies, ratings):
    """
    Prompts the user to add a new movie if it's not in the dataset.
    """
    print("Movie not found. Please add the movie details:")
    title = input("Enter the movie title: ")
    genres = input("Enter genres (comma-separated): ").split(',')  #The function lets the user add a new movie and their rating if the movie isn't already in the dataset, then updates the movies and ratings datasets with this new data.
    overview = input("Enter a short overview: ")
    keywords = input("Enter keywords (comma-separated): ").split(',')
    cast = input("Enter main cast (comma-separated): ").split(',')
    crew = input("Enter director(s) (comma-separated): ").split(',')
    user_rating = float(input("Rate this movie (1-5): "))

    # Add movie to dataset
    new_movie_id = movies['movie_id'].max() + 1
    new_movie = pd.DataFrame({
        'movie_id': [new_movie_id],
        'title': [title],
        'overview': [overview],
        'genres': [genres],
        'keywords': [keywords],
        'cast': [cast],
        'crew': [crew]
    })
    movies = pd.concat([movies, new_movie], ignore_index=True)

    # Add user rating to ratings
    new_rating = pd.DataFrame({
        'userId': [ratings['userId'].max() + 1],
        'movieId': [new_movie_id],
        'rating': [user_rating],
        'timestamp': [pd.Timestamp.now().timestamp()]
    })
    ratings = pd.concat([ratings, new_rating], ignore_index=True)

    return movies, ratings

```

```python
# Step 8: Hybrid Recommendation System (Combining Content-Based and Collaborative Filtering)
def hybrid_recommend(user_id, movie):
    """
    Hybrid recommendation system combining content-based and collaborative filtering.
    """
    if movie in new_df['title'].values:
        # Get content-based recommendations
        content_results = content_recommend(movie) #Content-based recommendations are based on the similarity between movies (e.g., based on genres, cast, keywords).
    else:
        print(f"'{movie}' not found. Adding new movie...")
        global movies, ratings
        movies, ratings = add_new_movie(movies, ratings)
        content_results = content_recommend(movie)

    # Get collaborative filtering recommendations
    collaborative_results = collaborative_recommend(user_id) # Collaborative filtering is based on user behavior and the ratings of other users

    # Ensure both results are not empty
    if not content_results:
        content_results = [("No content-based recommendations available", 0)]
    if not collaborative_results:
        collaborative_results = [("No collaborative recommendations available", 0)]

    # Combine recommendations
    hybrid_results = content_results[:3] + collaborative_results[:2]
    return hybrid_results
```

```python
# Step 9: Search and Recommend Function
def search_and_recommend():
    """
    Allows user to input a movie name and fetches hybrid recommendations.
    """
    user_id = int(input("Enter your User ID: "))
    while True:
        movie = input("Enter a movie name (or type 'exit' to quit): ")
        if movie.lower() == 'exit':
            print("Exiting the recommendation system. Goodbye!")
            break
        recommendations = hybrid_recommend(user_id, movie)

        print(f"\nHybrid Recommendations for '{movie}':")
        for title, score in recommendations:
            print(f"{title} (Score: {score:.2f})")

# Run the recommendation system
search_and_recommend()
```
